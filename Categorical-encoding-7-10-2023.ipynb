### drug dataset

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("drug200.csv")
df
```

```python
label_encoder = LabelEncoder() 
df['sex_labelencoded'] = label_encoder.fit_transform(df['Sex'])
df['BP_labelencoded'] = label_encoder.fit_transform(df['BP'])
df['Cholesterol_labelencoded'] = label_encoder.fit_transform(df['Cholesterol'])
df
```

```python
df['Drug'].value_counts()
```

```python
df.columns
```

```python
df_encoded = pd.get_dummies(df, columns=['Drug'])

df_encoded
```

```python
convert_to_int=['Drug_DrugY','Drug_drugA','Drug_drugB','Drug_drugC','Drug_drugX']
df_encoded[convert_to_int] = df_encoded[convert_to_int].astype(int)
df_encoded
```
